## 3.1 조건문 (if 문)

참과 거짓을 판단하는 문장

- 조건이 참일 경우 실행
- 조건이 거짓일 경우 pass
- 들여쓰기(indent) 주의
- 콜론 주의

```python
a, b = map(int, input().split())

if a>b:
    print(f'{a} is bigger than {b}')
    print(f'(Output success!)')
elif a<b:
    print(f'{a} is smaller than {b}')
    print('(Output success!)')
else:
    print('(Error... Try again...)')
```

## 3.2 반복문 -1 (while 문)

조건을 만족하는 한, 반복적으로 실행시키는 문장

(! 무한반복되지 않도록 주의)

```python
count = 0
while count<10:
    print(f'count : {count}')
    count += 1
```

### 3.2.1 break문

```python
a=10; count=0
while a>3:
    count += 1
    a += 1
    print(f'a = {a}\tcount = {count}')

    # while loop counting : 10번만 딱 돌겠다!!
    if count>10:
        break
print('\nEnd.')
```

### 3.2.2 continue문

```python
count=0
while True:
    count += 1

    if count%2 == 1:
        # continue : 아래 있는 코드를 무시하고 다시 while 조건으로 간다.
        continue
        print(f'반복횟수 = {count}')
    if count >= 10:
        print('End.')
        break
```

# 4.표현식(Comprehension)

다른 언어에서는 잘 볼 수 없는 파이썬만의 편리한 문법

list, set, dictionary 안에서 for문과 if문을 사용하여, 원소들을 구성하는 방법

## 4.1 if Comprehension

```python
score = 60

# original if loop
if score >= 80:
    pf = 'pass'
else:
    pf = 'fail'
print(pf)

# if loop comprehension
pf = 'pass' if score>=80 else 'fail'
print(pf)

# elif --> 너무 길어지면 가독성 별루...
```

## 4.2 List Comprehension

- [item **for** item **in** iterable]
- [item **for** item **in** iterable **if** condition]

```python
sentence = "조건 판단 후, 참/거짓에 따라 다른 명령을 실행하기 위하여 \
사용하는 것을 if문이라고 한다."
words = sentence.split()
print(words)
```

```python
# 각 value의 길이를 알고 싶다.
words_len = [len(word) for word in words]
print(words_len)
```

```python
# 조건 추가 : length 3 이상만 추출
words_3 = [word for word in words if len(word) > 2]
print(words_3)

# original version
words_3_v2 = []
for word in words:
    if len(word) > 2:
        words_3_v2.append(word)
print(words_3_v2)
```

## 4.3 Set Comprehension (updated at python3)

- {item **for** item **in** iterable}
- {item **for** item **in** iterable **if** condition}

```python
sentence = "조건 판단 후, 참/거짓에 따라 다른 명령을 실행하기 위하여 \
사용하는 것을 if문이라고 한다."
words = sentence.split()
print(words)
```

```python
length = {len(word) for word in words}
print(length)

# set으로 바꿔줘도 큰 차이 없음...????? 머선말이고 ㅠㅠ
length = set([word for word in words])
print(length)
```

## 4.4 Dictionary Comprehension (updated at python3)

- {key : value **for** item **in** iterable}
- {key : value **for** item **in** iterable **if** condition}

```python
country_capital = {'대한민국':'서울', '영국':'런던', '중국':'북경', 
                   '일본':'동경'}
country_capital
```

```python
'''
바꿔서 출력하고 싶음 --> capital_country
일일이? 일일히? 다 하면 귀찮쓰
'''
capital_country = {capital:country for country, capital 
                   in country_capital.items()}
capital_country
```

```python
# 이건 또 머선말이구 ㄷㄷ
capital_country = {country_capital[country]:country for country 
                   in country_capital.keys()}
capital_country
```

# 5.자료형과 제어문 활용 예제

```python
# 1. 홍길동 평균 점수
# 국어 80, 영어 75, 수학 55

# solution1 : 새로운 변수 추가 시 복잡해짐.
kor=80; eng=75; math=55
sum = kor + eng + math
avg = sum/3
print(f'홍길동의 평균 점수 : {avg}')

# solution2 : 새로운 변수 추가 시 편리함.
hong = {'국어':80, '영어':75, '수학':55, '도덕':80}
sum = 0
for score in hong.values():
    sum += score
avg = sum/len(hong)
print(f'홍길동의 평균 점수 : {avg}')
```
