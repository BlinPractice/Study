```python
# import numpy as np
# numpy 패키지를 불러올거에요! as np --> np라고 별명을 지어줄거에요~
# import sklearn.metrics as metrics

import numpy as np
import sklearn.metrics as metrics
```

```python
# 0 : '암이 아니다' 라고 예측 
# 1 : '암이다' 라고 예측

# y : 실제 정답
# p : 예측하는 값 (모델)

y = np.array([1,1,1,1,0,0]) # 실제 답
p = np.array([1,0,1,1,0,0]) # 나의 예측값(모델)
```

```python
# 첫번째 방법 : 직접 구하는 방법 :)

# 정확도(acc)
accuracy = np.mean(np.equal(y,p))
right = np.sum(y * p == 1)

# 정밀도(precision)
precision = right / np.sum(p)

# 재현율(recall)
recall = right / np.sum(y)

# f1-score
f1 = 2 * precision*recall / (precision + recall)
```

```python
print('accuracy : ', accuracy)
print('precision : ', precision)
print('recall : ', recall)
print('f1-score : ', f1)
```

```python
# 2. 두번째 방법 : 사이킷런 패키지를 이용해서 구하는 방법

print('accuracy : ', metrics.accuracy_score(y, p))
print('precision : ', metrics.precision_score(y, p))
print('recall : ', metrics.recall_score(y, p))
print('f1-score : ', metrics.f1_score(y, p), '\n')

print(metrics.classification_report(y,p))
print(metrics.confusion_matrix(y,p))
```
