# 0. Types of features in Titanic

    1. Categorial : A collection of information that is divided into groups.
        ex) Embarked (Cherbourg : C, Queentown : Q, Southampton : S)

    2. Ordinal : Similar to categorial but they have an order.
        ex) Pclass (1, 2, 3)

    3. Binary : A categorial feature which has only 2 types of categories. Which is often represted as 0 and 1.
        ex) Sex (Male / Female)

    4. Continuous : They can take up any value between the min and max values in a column.
        ex) Age, Fare

    5. Count : Represent the count of a variable.
        ex) Sibling and Spouse (SibSP), Parents and Children (Parch)

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

- All the yellow lines that we see are missing values.
- We can see that majority of values are missing in cabin and age.

- These are the following missing values.

```python
train.head()
```

```python
train.drop(['Name', 'Ticket', 'Cabin', 'Fare'], axis=1, inplace=True)
train
```

```python
# relationship between Pclass and Age
sns.boxplot(x='Pclass', y='Age', data=train)
```

```python
# Filling up mean values for null
def impute_age_train(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
def impute_age_test(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 42
        elif Pclass == 2:
            return 27
        else :
            return 25
    else:
        return Age
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(impute_age_train, axis=1)
train['Embarked'].describe()
```

- S is common in Embarked. We will fill null values with it.

```python
common = 'S'
train['Embarked'] = train['Embarked'].fillna(common)

train.isnull().sum()
```

# Categorical features

```python
Embarked = pd.get_dummies(train['Embarked'], drop_first=True)
Embarked.head()
```

```python
# converting values
train['Sex'] = train['Sex'].map({'male':0, 'female':1})
train.drop(['Embarked'], axis=1, inplace=True)

train
```

```python
train = pd.concat([train, Embarked], axis=1)
train.head()
```

```python
train.loc[train['Age'] <= 16, 'Age'] = 0
train.loc[(train['Age'] > 16) & (train['Age'] <= 32), 'Age'] = 1
train.loc[(train['Age'] > 32) & (train['Age'] <= 48), 'Age'] = 2
train.loc[(train['Age'] > 48) & (train['Age'] <= 64), 'Age'] = 3
train.loc[ train['Age'] > 64, 'Age'] = 4
```

```python
train = train.drop('PassengerId', axis=1)
train
```
