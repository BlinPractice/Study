### 예외처리

```python
number = 10

# 예외처리를 적용하지 않은 경우
print(number/0)     # 정수를 0으로 나누는 연산은 연산 불가능하므로 예외 발생
```

```python
# 예외처리를 적용한 경우
try:
    print(number/0)     # 예외 발생 가능성이 있는 코드 블럭
except Exception as e:  # 예외 발생 시 처리할 부분
    print(e)
finally:    # 예외 발생과 관련 없이 무조건 실행되는 부분
    print('stop script...')
```

```python
try:
    print(number/0)     # 예외 발생 가능성이 있는 코드 블럭
except Exception as e:  # 예외 발생 시 처리할 부분
    print(number/1)
finally:    # 예외 발생과 관련 없이 무조건 실행되는 부분
    print('stop script...')
```

```python
# 예외가 발생하지 않은 경우
try:
    print(number/10)     # 예외 발생 가능성이 있는 코드 블럭
except Exception as e:  # 예외 발생 시 처리할 부분
    print(number/1)
finally:    # 예외 발생과 관련 없이 무조건 실행되는 부분
    print('stop script...')
```
